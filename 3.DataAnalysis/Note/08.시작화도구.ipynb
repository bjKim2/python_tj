```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head(2)
```

```python
df.tail()
```

```python
df.info()
```

```python
# NaN을 전(Prev) 데이터로 채움
df = df.fillna(method='ffill') # NaN을 가장 근처값으로채움 (여기선 전 행 데이터)
df.head()
```

```python
df.info() # NaN 데이터 확인
```

```python
df['전출지별'].isnull().describe()
```

```python
df['전출지별']
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
# 전출지별이 서울특별시 이고 전입지별이 서울특별시가 아닌곳
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df.loc[mask,:]
```

```python
# 전출지별 컬럼 삭제
df_seoul.drop('전출지별',axis=1,inplace=True)
df_seoul.head(2)
```

```python
# 전입지별 -> 전입지 로 컬럼 이름 변경
df_seoul.rename(columns={'전입지별':'전입지'},inplace=True)
df_seoul
```

```python
df_seoul
```

```python
# 전입지를 index로 변경하기 
df_seoul.set_index("전입지", inplace=True)
df_seoul.head()
```

```python

df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택 
sr_one =df_seoul.loc['경기도',:]
sr_one
```

```python
sr_one
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 차트 제목 추가
plt.plot(sr_one)
plt.title('서울 -> 경기 이동 인구')

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```


## 그래프에 Style 적용하기

```python
# 그래프 꾸미기

# Style 서식 지정
plt.style.use('ggplot')



# 그림 사이즈  지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))

# x축 눈금 라벨 회전하기
plt.xticks(rotation='vertical')

#y축 눈금 라벨 회전하기
plt.yticks(rotation=45)

plt.title('서울 -> 경기 인구 이동',size = 20)
plt.xlabel('기간', size = 20)
plt.ylabel('이동 인구수', size = 20)
plt.plot(sr_one)

# Marker표시
plt.plot(sr_one, marker = 'o', markersize = 10)

# y축 범위 지정
plt.ylim(50000,800000) # 최소값, 최대값 (시작과 끝이 아님)

# 화살표 표시
plt.annotate(
                '',
                xy = (20,620000), # 화살표의 머리부분
                xytext=(2,290000), # 화살표의 꼬리
                # xycoords='data', '# 좌표체계
                arrowprops=dict(arrowstyle='->', color='skyblue', lw=5)
            )

plt.annotate(
    '인구 이동 증가(1970~1995)', # 텍스트
    xy = (10, 400000), # 텍스트 위치 기준점
    rotation = 24, # 텍스트 회전 각도
    va = 'baseline', # 텍스트 상하 정렬
    ha = 'center', # 텍스트 좌우 정렬
    fontsize = 12 # 텍스트 크기
)

plt.annotate(
                '',
                xy = (47,450000), # 화살표의 머리부분
                xytext=(30,580000), # 화살표의 꼬리
                # xycoords='data', '# 좌표체계
                arrowprops=dict(arrowstyle='->', color='green', lw=5)
            )

plt.annotate(
    '인구 이동 감소(1995~2017)', # 텍스트
    xy = (39, 540000), # 텍스트 위치 기준점
    rotation = -10, # 텍스트 회전 각도
    va = 'baseline', # 텍스트 상하 정렬
    ha = 'center', # 텍스트 좌우 정렬
    fontsize = 12 # 텍스트 크기
)
# 범례 표시
plt.legend(labels=['서울->경기'], loc = 'best', fontsize = 15)

# 이미지.저장
plt.savefig("../Data/seoul.png")

plt.show()

```

## 화면을 분할하여 그래프 여러개 그리기

```python
#Warning 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# 그래프 꾸미기

# Style 서식 지정
plt.style.use('ggplot')

# 그래프 객체 생성(figure에 2개의 서브플폿 생성)
fig = plt.figure(figsize=(1,14))
ax1 = fig.add_subplot(2,1,1) # 2행 1열의 1번째 
ax2 = fig.add_subplot(2,1,2) # 2행 1열의 2번째 

ax1.plot(sr_one,'o',markersize = 10)
ax2.plot(sr_one,marker='.',markerfacecolor= 'green', markersize= 10, # o,*,+,|,.
            color= 'olive', linewidth=2,label = '서울->경기') # label 은 범례가 있어야 나옴

ax2.legend(loc='best')
ax2.set_title('서울 -> 경기 인구 이동',size=20)

ax1.set_ylim(50000,800000)
ax2.set_ylim(50000,800000)

ax1.set_xticklabels(sr_one.index, rotation = 75)
ax2.set_xticklabels(sr_one.index, rotation = 45)

plt.show()

```

### 동일한 그림에 여러 개의 그래프를 추가

```python
df_seoul
```

```python
# 서울에서 '충청남도', '경상북도' , '강원도'로 이동한 인구데이터 값만 선택
df_3 = df_seoul.loc[['충청남도','경상북도','강원도']]
df_3
```

```python
# Style
plt.style.use('ggplot')

fig = plt.figure(figsize= (20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(
        df_3.columns,
        df_3.loc['충청남도'],
        marker = 'o',
        markerfacecolor = 'green',
        markersize = 10,
        color = 'olive',
        linewidth = 2,
        label = '서울 -> 충남'
        )

ax.plot(
        df_3.columns,
        df_3.loc['경상북도'],
        marker = 'o',
        markerfacecolor = 'blue',
        markersize = 10,
        color = 'skyblue',
        linewidth = 2,
        label = '서울 -> 경북'
        )
        
ax.plot(
        df_3.columns,
        df_3.loc['강원도'],
        marker = 'o',
        markerfacecolor = 'red',
        markersize = 10,
        color = 'red',
        linewidth = 2,
        label = '서울 -> 강원'
        )



ax.legend(loc = 'best')
ax.set_title('서울->충남, 경북, 강원 인구 이동', size = 20)

ax.set_xlabel('기간',size = 12)
ax.set_ylabel('이동 및 인구수', size = 12)

ax.set_xticklabels(df_3.columns,rotation= 90)

# 축 눈금 라벨 크기
ax.tick_params(axis='x', labelsize = 9)
ax.tick_params(axis='y', labelsize = 5)

plt.show()
```

```python
# 서울에서 '충청남도', '경상북도' , '강원도'로 이동한 인구데이터 값만 선택
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도']]
df_4
```

```python
# Style

plt.style.use('ggplot')

fig = plt.figure(figsize = (20,14))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(
    df_4.columns,
    df_4.loc['충청남도'],
    marker = 'o',
    markerfacecolor = 'green',
    color = 'olive',
    label = '서울 -> 충남'
)
ax1.set_title('서울 -> 충남 인구 이동',size=20)
ax1.set_ylim(1,70000)
ax1.set_xticklabels(df_4.columns,rotation = 75)

ax2.plot(
    df_4.columns,
    df_4.loc['경상북도'],
    marker = 'o',
    markerfacecolor = 'blue',
    color = 'blue',
    label = '서울 -> 경북'

)
ax2.set_title('서울 -> 경북 인구 이동',size=20)
ax2.set_ylim(1,70000)
ax2.set_xticklabels(df_4.columns,rotation = 75)

ax3.plot(
    df_4.columns,
    df_4.loc['강원도'],
    marker = 'o',
    markerfacecolor = 'red',
    color = 'red',
    label = '서울 -> 강원'
)
ax3.set_title('서울 -> 강원 인구 이동',size=20)
ax3.set_ylim(1,70000)
ax3.set_xticklabels(df_4.columns,rotation = 75)

ax4.plot(
    df_4.columns,
    df_4.loc['전라남도'],
    marker = 'o',
    markerfacecolor = 'orange',
    color = 'orange',
    label = '서울 -> 전남'
)
ax4.set_title('서울 -> 전남 인구 이동',size=20)
ax4.set_ylim(1,70000)
ax4.set_xticklabels(df_4.columns,rotation = 75)

ax1.legend(loc = 'best')
ax2.legend(loc = 'best')
ax3.legend(loc = 'best')
ax4.legend(loc = 'best')

plt.show()
```

### 2010~2017년의 기간동안 서울에서 각 시도로 이동한 인구의 합계를 구하여 그래프로 비교

```python
# 서울에서 충청남도, 경상북도, 강원도, 전라남도로 이동한 인구 데이터 값만 선택
col_years = list(map(str, range(2010,2018)))
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'],col_years]
df_4
```

```python
df_4['합계'] = df_4.sum(axis=1)
```

```python
df_total = df_4[['합계']].sort_values(by = '합계',ascending=True)
df_total
```

```python
plt.style.use('ggplot')
df_total.plot(
    kind = 'barh',  # barh 는 옆으로 그리는 막대 
    color = 'cornflowerblue',
    figsize = (10,5)
    
    
)
plt.title('서울 -> 타시도 인구 이동')
plt.ylabel('전입지')
plt.xlabel('이동 인구수')

plt.show()
```
