```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 2개의 축을 갖는 그래프 작성
- 남북한 발전량 데이터 셋을 사용
- 막대그래프는 북한의 발전량을 표시
- 선그래프는 북한 발전량의 전년 대비 증감률


```python
df = pd.read_excel("../Data/남북한발전전력량.xlsx")
df.head()
```

```python
# 북한 전력량만 선택
df = df.loc[5:9]
df
```

```python
df.columns
```

```python
# 전력량 컬럼 삭제
df.drop("전력량 (억㎾h)", axis=1, inplace=True)
df
```

```python
# index를 발전 전력별로 설정
df.set_index('발전 전력별',inplace= True)
df
```

```python
df = df.T
df.head()
```

```python
# 합계 컬럼을 총발전량으로 변경
df = df.rename(columns={'합계':'총발전량'})
df.head()
```

```python
# 증감률(변동률) 계산
# 총발전량 -1 년 이라는 컬럼을 생산
df['총발전량-1년'] = df['총발전량'].shift(1)
df.head()
```

```python
# 증감률 컬럼 생성
df['증감률'] = ((df['총발전량']/df['총발전량-1년'])-1)*100
df.head()
```

```python
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 2축 그래프 그리기
plt.style.use('ggplot')

ax1 = df[['수력','화력']].plot(kind = 'bar',figsize = (20,10),stacked=True)
ax2 = ax1.twinx() # x 축을 ax1과 같게 설정
ax2.plot(df.index,df.증감률, ls = '--', marker = 'o',markersize=10,color='green',label = '전년대비 증감률(%)')



ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도',size = 20)
ax1.set_ylabel('발전량 (억 kwh)',size = 20)
ax2.set_ylabel('전년대비 증감률(%)')

plt.title("북한 전력 발전량(1990~2016)",size = 30)
ax1.legend(loc= 'upper left',fontsize = 30)
ax2.legend(loc= 'best',fontsize = 30) # best는 알아서하는 것

plt.show()
```
