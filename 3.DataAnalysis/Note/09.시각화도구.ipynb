### 2개의 축을 갖는 그래프 작성
- 남북한 발전량 데이터 셋을 사용
- 막대그래프는 북한의 발전량을 표시
- 선그래프는 북한 발전량의 전년 대비 증감률


```python
df = pd.read_excel("../Data/남북한발전전력량.xlsx")
df.head()
```

```python
# 북한 전력량만 선택
df = df.loc[5:9]
df
```

```python
df.columns
```

```python
# 전력량 컬럼 삭제
df.drop("전력량 (억㎾h)", axis=1, inplace=True)
df
```

```python
# index를 발전 전력별로 설정
df.set_index('발전 전력별',inplace= True)
df
```

```python
df = df.T
df.head()
```

```python
# 합계 컬럼을 총발전량으로 변경
df = df.rename(columns={'합계':'총발전량'})
df.head()
```

```python
# 증감률(변동률) 계산
# 총발전량 -1 년 이라는 컬럼을 생산
df['총발전량-1년'] = df['총발전량'].shift(1)
df.head()
```

```python
# 증감률 컬럼 생성
df['증감률'] = ((df['총발전량']/df['총발전량-1년'])-1)*100
df.head()
```

```python
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 2축 그래프 그리기
plt.style.use('ggplot')

ax1 = df[['수력','화력']].plot(kind = 'bar',figsize = (20,10),stacked=True)
ax2 = ax1.twinx() # x 축을 ax1과 같게 설정
ax2.plot(df.index,df.증감률, ls = '--', marker = 'o',markersize=10,color='green',label = '전년대비 증감률(%)')



ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도',size = 20)
ax1.set_ylabel('발전량 (억 kwh)',size = 20)
ax2.set_ylabel('전년대비 증감률(%)')

plt.title("북한 전력 발전량(1990~2016)",size = 30)
ax1.legend(loc= 'upper left',fontsize = 30)
ax2.legend(loc= 'best',fontsize = 30) # best는 알아서하는 것

plt.show()
```

### 히스토그램(histogram)
: 단위별(컬럼별) 데이터의 빈도수를 그래프로 표현

```python
# Style 서식 지정
plt.style.use('classic')
```

```python
df = pd.read_csv("../Data/auto-mpg.csv", header = None)
df.head()
```

```python
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
df.head()
```

```python
# 연비(mpg) 열에 대한 히스토그램 그리기
df['mpg'].plot(
    kind = 'hist',
    bins = 10, # 전체 등분 수 
    color = 'coral',
    figsize = (10,5),
)
# 그래프 그리기
plt.title("Histogram")
plt.xlabel('mpg')

plt.show()
```


- 그림만 그리면 무조건 코멘트가 있어야 됨. 그림에서 의미를 찾아내야됨.  
> 자동차의 연비값은 대부분 좌측편향이 되어 연비가 낮은 구간에 집중되어 있는 경향을 보인다.
- 통계의 해석이 실제값이 아니라 예측값이라 경향을 보인다.라는 말을 잘 씀

### 산점도 (Scatter plot)
- 서로 다른 두 변수 사이의 관계를 나타낸다.


```python
plt.style.use('default')
```

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도 그리기
df.plot(
    kind = 'scatter',
    x = 'weight',
    y = 'mpg',
    c = 'coral', # c 는 color, 바 플랏은 c로 적으면 안됨 
    s = 10, # s 는 점의 크기,
    figsize = (10,5)
)
plt.title('Scatter plot = mpg vs. weight')
plt.show()
```

- 산점도는 상관을 보기 때문에 역상관 상관 단어를 쓰는 게 좋다.
> 자동차 무게가 높을수록 연비는 낮아지는 역상관을 표현하고 있다.(쌤이 쓴거)  
> 연비는 차중에 역상관을 보인다. (내가쓴거)  
> 차중이 커질수록 연비의 분산이 작다.   

```python
# Cylinder 개수의 상대적 비율을 계산하여 시리즈 생성
cylinder_size = df.cylinders / df.cylinders.max() * 300 # 300 Hyper prameter, 기준점은 그대로라는 특징
print(cylinder_size)
```

```python
df.columns
```

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도 그리기
df.plot(
    kind = 'scatter',
    x = 'weight',
    y = 'mpg',
    c = 'coral', # c 는 color, 바 플랏은 c로 적으면 안됨 
    s = cylinder_size, # s 는 점의 크기,
    figsize = (10,5)
)
plt.title('Scatter plot = mpg vs. weight')
plt.show()
```

### 버블(Bubble) Chart

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도 그리기
df.plot(
    kind = 'scatter',
    x = 'weight',
    y = 'mpg',
    c = 'blue', # c 는 color, 바 플랏은 c로 적으면 안됨 
    s = cylinder_size, # s 는 점의 크기,
    alpha = 0.3,
    figsize = (10,5)
)
plt.title('Scatter plot = mpg vs. weight')
plt.show()
```

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도 그리기
df.plot(
    kind = 'scatter',
    x = 'weight',
    y = 'mpg',
    c = cylinder_size, # c 는 color, 바 플랏은 c로 적으면 안됨 
    s = 50, # s 는 점의 크기,
    alpha = 0.3,
    figsize = (10,5),
    marker = '+',
    cmap = 'viridis'
)
plt.title('Scatter plot = mpg vs. weight')
plt.show()
```

### 파이차트(pie chart)
- 원을 파이처럼 나누어서 표현
- 조각의 크기는 해당변수에 속하는 데이터 값의 크기에 비례(%)


#### 자동차의 국가별 데이터 갯수를 파이차트로 표현

```python
plt.style.use('default')
```

```python
df.head()
```

```python
# 국가별 Count 계산
df['count'] = 1
df.head()
```

```python
# origin열을 기준으로 그룹화, 합계 연산
df_origin = df.groupby('origin').sum() # aggregate 랑 똑같네
df_origin.head()
```

```python
# index값을 변경
df_origin.index = ['USA', 'EU', 'JAPAN']
df_origin
```

```python
# 파이차트 그리기 
df_origin['count'].plot(
    kind = 'pie',
    figsize = (7,5),
    autopct = '%.2f%%', #autopercent , %%를 두개써야 %모양이 나옴
    startangle = 90, # 시작 각도
    colors = ['chocolate', 'bisque', 'cadetblue'],
    
)

plt.title("Model Origin", size = 20)
plt.axis('equal') # 파이 차트가 만약에 찌그러진다면 써야되는 문구
plt.legend(labels = df_origin.index, loc = 'lower right')
plt.ylabel('')
plt.show()
```

### 박스플롯(boxplot) 
- 범주형 데이터의 분포를 파악
- 상자수염 이라고도 함

##### 제조 국가별 연비 분포 박스플롯

```python
plt.style.use('seaborn-poster')
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 그래프 객체 생성
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# 수직 박스 플롯
ax1.boxplot(
    x = [
        df[df['origin'] == 1]['mpg'],
        df[df['origin'] == 2]['mpg'],       
        df[df['origin'] == 3]['mpg']
    ],
    labels = df_origin.index
)

ax2.boxplot(
    x = [
        df[df['origin'] == 1]['mpg'],
        df[df['origin'] == 2]['mpg'],       
        df[df['origin'] == 3]['mpg']
    ],
    labels = df_origin.index,
    vert = False
)

ax1.set_title('제조 국가별 연비 분포(수직 박스 플롯)',size = 15)
ax2.set_title('제조 국가별 연비 분포(수평 박스 플롯)',size = 15)

plt.show()
```
