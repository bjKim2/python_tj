```python
bookContents = []

for url in hyperlink_lists:
    res = req.urlopen(url)
    soup = BeautifulSoup(res,"html.parser")
    book_contents = soup.select_one(".folded")
    bookContents.append(book_contents.text)
```

```python
ridibooks['Contents'] = pd.Series(bookContents)
ridibooks.head()
```

```python
titles = []
contents = []
for i in range(1,8):
    url = 'https://ridibooks.com/bestsellers/general?order=monthly&page='+str(i)
    soup = BeautifulSoup(req.urlopen(url), "html.parser")
    scraping_lists = soup.select(".title_text")
    for title in scraping_lists:
        titles.append(title.string.strip())
    if(i == 1):
        titles.pop(0)
    
    s1 = soup.select(".title_link")
    if(i == 1):
        s1.pop(0)
    hyperlink_lists = []
    for i in range(len(s1)):
        hyperlink_lists.append('https://ridibooks.com'+s1[i].attrs['href'])

    for url in hyperlink_lists:
        res = req.urlopen(url)
        soup = BeautifulSoup(res,"html.parser")
        book_contents = soup.select_one(".folded")
        contents.append(book_contents.text)
```
