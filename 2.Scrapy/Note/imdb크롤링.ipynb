```python
driver = webdriver.Chrome( ChromeDriverManager().install(), options=chrome_options)
driver.get("https://www.imdb.com/chart/top/?sort=us,desc&mode=simple&page=1")
```

```python
page_source = driver.page_source 
soup = BeautifulSoup(page_source, "html.parser")
movie_list = soup.select('#main > div > span > div > div > div.lister > table > tbody > tr > td.titleColumn > a')
```

```python
movie_list = movie_list[:100]
```

```python
movie_id = []
for i in range(len(movie_list)):
    movie_id.append(movie_list[i].attrs['href'].split('/')[2])
```

영화 리뷰 100개 씩 열기

```python
movie_name = []
content = []
title = []
score = []
max_score = []
date = []
vote = []
max_vote = []
```

```python
driver = webdriver.Chrome( ChromeDriverManager().install(), options=chrome_options)
for j in range(len(movie_id)):
    print(j,'번',movie_list[j].text)
    for str01 in ['asc','desc']:
        url = 'https://www.imdb.com/title/'+ movie_id[j]+'/reviews?sort=userRating&dir='+ str01 +'&ratingFilter=0' ## 별점 낮은 순 
        driver.get(url)
        time.sleep(5)
        for k in range(3):
            xpath = '//*[@id="load-more-trigger"]'
            element_sel_gu = driver.find_element(By.XPATH,xpath).click()
            time.sleep(3)
            print('버튼',k+1)
        page_source = driver.page_source 
        soup = BeautifulSoup(page_source, "html.parser")
        
        
        all_review = soup.select("#main > section > div.lister > div.lister-list > div > div.review-container > div.lister-item-content")
        temp_vote = soup.select('div.actions.text-muted')
        for i in range(len(all_review)):
            movie_name.append(movie_list[j].text.strip())
            content.append(all_review[i].find(attrs = {'class':'text'}).text.strip())
            title.append(all_review[i].find(attrs = {'class':'title'}).string.strip())
            score.append(all_review[i].find('span').find_all('span')[0])
            max_score.append(all_review[i].find('span').find_all('span')[1])
            date.append(all_review[i].find(attrs = {'class':"review-date"}).string.strip())
            # all_review[0].find(attrs = {'class':"actions text-muted"}).text.strip().split('\n')[0]
            if re.findall('\d+',temp_vote[i].text.strip().split('\n')[0]) == None:
                vote.append(0)
                max_vote.append(0)
            else:
                vote.append( re.findall('\d+',temp_vote[i].text.strip().split('\n')[0])[0])
                max_vote.append( re.findall('\d+',temp_vote[i].text.strip().split('\n')[0])[1])
            # vote.append( all_review[i].find(attrs = {'class':"actions text-muted"re.findall('\d+',}).text.strip().split('\n')[0])[0])
            # max_vote.append(re.findall('\d+', all_review[i].find(attrs = {'class':"actions text-muted"}).text.strip().split('\n')[0])[1])
        
    
    
    
    
```

```python
print(len(movie_name),
len(content),
len(title),
len(score),
len(max_vote),
len(vote))
```

```python
df = pd.DataFrame({'movie':movie_name,'content':content, 'title':title, 'score':score, 'max_score':max_score, 'date':date, 'vote':vote, 'max_vote':max_vote})
df
```

```python
df2 = df.copy()
```

```python
for i in range(len(df)):
    df2.loc[i,'score'] = df.loc[i,'score'].text
    df2.loc[i,'max_score'] = df.loc[i,'max_score'].text[1:]
```
