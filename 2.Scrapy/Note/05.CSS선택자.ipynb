```python
from bs4 import BeautifulSoup
import urllib.request as req

from urllib import parse

auth = "저자"
name = "윤동주"


url = "https://ko.wikisource.org/wiki/"
url = url + parse.quote(auth) + ":" + parse.quote(name)
#한글 검색은 꼭 파싱을 하자!

print(url)

res = req.urlopen(url)
soup = BeautifulSoup(res,"html.parser")
```

```python
books = soup.select("#mw-content-text > div.mw-parser-output > ul > li > b > a")
print(books)
#mw-content-text > div.mw-parser-output > ul:nth-child(6) > li > b > a
#mw-content-text > div.mw-parser-output > ul:nth-child(6) > li > a
```

```python
books2 = soup.select("#mw-content-text > div.mw-parser-output > ul > li > ul > li > a")
for i in books2:
    print(i.string)
```

```python
books3 = soup.select("#mw-content-text > div.mw-parser-output > ul > li a")
for i in books3:
    if(i.string == '증보판'):
        continue
    print("-",i.string)
```
