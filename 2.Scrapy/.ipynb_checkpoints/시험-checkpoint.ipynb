```python
# Library
from bs4 import BeautifulSoup
import urllib.request as req

# Site Address
url = "https://news.daum.net/"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup으로 분석하기

soup = BeautifulSoup(res,"html.parser")
print(soup)
```

```python
temp= soup.select("body > div.container-doc > main > section > div > div.content-article > div.box_g.box_news_issue > ul > li > div > div > strong > a")
titles = []
urls = []
for i in temp:
    titles.append(i.string.strip())
    urls.append(i['href'])

print(titles)
print(urls)
```

```python
import pymysql

## Insert Data #1
# connection
conn = pymysql.connect(host="127.0.0.1", user = "root", password="qwer1234", \
                        db = "daum_news", charset="utf8")

# Connection으로 부터 Cursor 생성
curs = conn.cursor()
# SQL 문장
sql = "insert into news (titles,url) values (%s, %s)"
# Insert 실행
for i in range(len(titles)):
    curs.execute(sql, (titles[i],urls[i]))

conn.commit()

# SQL 문장
sql = "select * from news"
curs.execute(sql)
# Data Fetch
rows = curs.fetchall()
print(rows)
```
