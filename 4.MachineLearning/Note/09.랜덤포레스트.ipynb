```python
# 랜덤 포레스트 (ensenble기법)

import pandas as pd

#필기 쌤은 랜덤 포레스트로 먼저 머신러닝 한 뒤 그 것을 기준으로 함.
```

```python
# 붓꽃의 데이터 읽어 들이기
csv= pd.read_csv("../Data/iris.csv")
```

```python
csv.head()
```

```python
# 필요한 열 추출하기
csv_data = csv[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]
csv_label = csv['Name']
```

```python
# 학습 전용 데이터와 테스트 전용 데이터로 나누기 
from sklearn.model_selection import train_test_split

train_data, test_data, train_label, test_label = train_test_split(csv_data,csv_label)
```

```python
# 데이터 학습 시키고 예측하기
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(train_data,train_label)
```

```python
# 정확도
clf.score(test_data,test_label)
```

```python
# Confusion Matrix
from sklearn import metrics

pred = clf.predict(test_data)
print(metrics.classification_report(pred,test_label))
print(metrics.confusion_matrix(pred,test_label))
pd.DataFrame(metrics.confusion_matrix(pred,test_label),columns=(test_label.unique()+'(예측)'),index=csv['Name'].unique())
```

# Data 정제하기

```python
#연습하기
print(ord('a'))
print(chr(97))
```

```python
# 데이터 냅의 문자를 숫자로 변환하기
label = []
data = []

for row_index, row in mr.iterrows():
    # print(row_index,row)
    # print(row.loc[1:])
    label.append(row.loc[0])
    row_data = []
    for v in row.loc[1:]:
        row_data.append(ord(v))
    data.append(row_data)

print(data[0:20])
```

```python
# label을 DataFrame으로 만들기
labelTemp = pd.DataFrame(label)
labelTemp.head()
```

```python
# data를 DataFrame으로 만들기 Column Name을 겹치지 않게 하기
dataTemp = pd.DataFrame(data)
dataTemp.rename(columns=lambda x:x+1, inplace=True )
dataTemp.head()
```

```python
# 2개의 DataFrame을 합치기
mr2 = pd.concat([labelTemp,dataTemp],axis = 1) # axis='columns'도 같음
mr2.head()
```

```python
# RandomForest로 정확도 측정해 보기

clf = RandomForestClassifier()
```

```python
train_data, test_data, train_label, test_label  = train_test_split(dataTemp,labelTemp,stratify=labelTemp)
```

```python
clf.fit(train_data,train_label)

clf.score(test_data,test_label)
```

```python
pred = clf.predict(test_data)
print(metrics.classification_report(pred,test_label))
```

```python
clf.predict_proba(test_data)
```

### One Hot Encofing 실시
: 숫자 데이터가 숫자로서 의미가 있으면 상관없지만 위의 데이터는 분류를 위한 데이터 이므로 숫자 크기가 의미가 없다.   
이 때 사용하는 것이 One Hot Encoding 이다.

```python
# 1번열 data의 종류 확인
mr2[1].unique()
```

```python
# 연습하기
dataTemp2 = dataTemp[:] #backup
for i in range(1,23):
    dataTemp = pd.get_dummies(data= dataTemp, columns = [i], prefix=str(i))
dataTemp.head()
```

```python
train_data, test_data, train_label, test_label = train_test_split(dataTemp,labelTemp,stratify=labelTemp)
```

```python
clf.fit(train_data,train_label)
pred = clf.predict(test_data)
print(clf.score(test_data,test_label))
```
